```python
import pandas as pd
import calendar
```

```python
energyUnit = "kWh"
hoursinaday = 24

costColPostfix = "Cost"

optimization1 = "Prev"
cost1ColName = "{:}{:}".format(optimization1, costColPostfix)
util1ColName = "{:}{:}".format(cost1ColName, energyUnit)

optimization2 = "Prop"
cost2ColName = "{:}{:}".format(optimization2, costColPostfix)
util2ColName = "{:}{:}".format(cost2ColName, energyUnit)

savedColName = "Energy Cost Saved"
```

## Months

```python
months_costs_df = pd.read_csv("results/Months/Months-Costs/Months-Costs.txt", sep = ",")
months_costs_df.index += 1
months_costs_df.fillna(0, inplace = True)
months_costs_df = months_costs_df.astype({
    "Months":"float",
    "Simulations":"float",
    cost1ColName:"float",
    cost2ColName:"float",
})
months_costs_df[util1ColName] = round((months_costs_df[cost1ColName]/months_costs_df["Months"])/hoursinaday, 2)
months_costs_df[util2ColName] = round((months_costs_df[cost2ColName]/months_costs_df["Months"])/hoursinaday, 2)
months_costs_df[savedColName] = months_costs_df[cost1ColName]-months_costs_df[cost2ColName]
```

```python
months_costs_df
```

```python
months_costs_avg1 = round(months_costs_df[util1ColName].sum()/len(months_costs_df), 2)
months_costs_avg2 = round(months_costs_df[util2ColName].sum()/len(months_costs_df), 2)
print ("The energy used at the rate of {:} {:} when using {:}. optimization approach".format(months_costs_avg1, energyUnit, optimization1))
print ("The energy used at the rate of {:} {:} when using {:}. optimization approach".format(months_costs_avg2, energyUnit, optimization2))
```

```python
monthwise_costs_df = pd.read_csv("results/Monthwise/Monthwise-Costs/Monthwise-Costs.txt", sep = ",")
monthwise_costs_df.index += 1
monthwise_costs_df.fillna(0, inplace = True)
monthwise_costs_df = monthwise_costs_df.astype({
    cost1ColName:"float",
    cost2ColName:"float",
})
monthwise_costs_df[util1ColName] = round((monthwise_costs_df[cost1ColName]/1)/hoursinaday, 2)
monthwise_costs_df[util2ColName] = round((monthwise_costs_df[cost2ColName]/1)/hoursinaday, 2)
monthwise_costs_df[savedColName] = monthwise_costs_df[cost1ColName]-monthwise_costs_df[cost2ColName]

monthwise_costs_df_row = {
    monthwise_costs_df.columns[0]: "2014-2015",
    monthwise_costs_df.columns[1]: "All",
    monthwise_costs_df.columns[2]: 0,
    monthwise_costs_df.columns[3]: monthwise_costs_df[cost1ColName].sum(),
    monthwise_costs_df.columns[4]: monthwise_costs_df[cost2ColName].sum(),
    monthwise_costs_df.columns[5]: monthwise_costs_df[util1ColName].sum(),
    monthwise_costs_df.columns[6]: monthwise_costs_df[util2ColName].sum(),
    monthwise_costs_df.columns[7]: monthwise_costs_df[savedColName].sum(),
}
monthwise_costs_df = monthwise_costs_df.append(monthwise_costs_df_row, ignore_index = True)
```

```python
monthwise_costs_df
```

```python
monthwise_costs_avg1 = round(monthwise_costs_df[util1ColName].sum()/len(monthwise_costs_df), 2)
monthwise_costs_avg2 = round(monthwise_costs_df[util2ColName].sum()/len(monthwise_costs_df), 2)
print ("The energy used at the rate of {:} {:} when using {:}. optimization approach".format(monthwise_costs_avg1, energyUnit, optimization1))
print ("The energy used at the rate of {:} {:} when using {:}. optimization approach".format(monthwise_costs_avg2, energyUnit, optimization2))
```
